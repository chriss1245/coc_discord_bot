```python
import discord
from discord.ext import commands

class MyBot(commands.Bot):
    def __init__(self, token, prefix):
        super().__init__(command_prefix=prefix, intents=discord.Intents.default())
        self.token = token

    async def on_ready(self):
        print(f"Bot conectado como {self.user}.")

    @commands.command()
    async def ping(self, ctx):
        await ctx.send("Pong!")

    @commands.command()
    async def assign(self, ctx, rol: discord.Role, miembro: discord.Member):
        await miembro.add_roles(rol)
        await ctx.send(f"Se ha asignado el rol {rol} a {miembro}.")

    @commands.command()
    async def unassign(self, ctx, rol: discord.Role, miembro: discord.Member):
        await miembro.remove_roles(rol)
        await ctx.send(f"Se ha eliminado el rol {rol} de {miembro}.")

    @commands.command()
    async def message(self, ctx, canal: discord.TextChannel, *, mensaje: str):
        await canal.send(mensaje)

    async def on_command_error(self, ctx, error):
        if isinstance(error, commands.errors.CommandNotFound):
            await ctx.send("Comando desconocido.")
        elif isinstance(error, commands.errors.MissingRequiredArgument):
            await ctx.send("Faltan argumentos.")
        else:
            await super().on_command_error(ctx, error)

    def run(self):
        super().run(self.token)


# Configuración del bot
TOKEN = secrets["discord"]["token"]
PREFIX = "!"

# Crear bot y ejecutarlo
bot = MyBot(TOKEN, PREFIX)
```
